### Feature Engineering
Goal - Create new, meaningful features that will help in:

1. Better Visualization
2. Building Ml models later (churn, sales prediction, etc.)
3. Extracting business insights

We'll create:
1. Sales Buckets - Low/Medium/High value orders
2. Country Buckets - Top markets vs Others
3. Recency Buckets - Recent vd Old customers (based on DAYS_SINCE_LASTORDER)
4. Customer-Level Aggregates - Total and average sales per customer
5. Price and Discount Features - MSRP vs actual price
6. Encoded Deal size - Convert DEALSIZE into numeric

### Sales buckets (Low/Medium/High)
We'll use quantiles of SALES to define buckets.

```python
## Check quantiles of SALES
df_clean["SALES"].quantile([0.25,0.5,0.75])
```

```python
## We will split into 3 buckets
## Define thresholds using quantiles
q1 = df_clean["SALES"].quantile(0.33)
q2 = df_clean["SALES"].quantile(0.66)

q1, q2
```

```python
def sales_bucket(x):
    if x <= q1:
        return "Low"
    elif x <= q2:
        return "Medium"
    else:
        return "High"

df_clean["SALES_BUCKET"] = df_clean["SALES"].apply(sales_bucket)

df_clean["SALES_BUCKET"].value_counts()
```

### Insights
Now every order is tagged as low,medium or high value.

```python
plt.figure(figsize=(6,4))
df_clean["SALES_BUCKET"].value_counts().plot(kind="bar")
plt.title("Count of Orders by Sales Bucket")
plt.xlabel("Sales Bucket")
plt.ylabel("Number of Orders")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Total sales by country
country_sales = (
    df_clean.groupby("COUNTRY")["SALES"]
    .sum()
    .sort_values(ascending=False)
)

country_sales.head(10)
```

```python
# suppose we top 5 as key markets
top_countries = country_sales.head(5).index.tolist()
top_countries
```

```python
def country_bucket(x):
    if x in top_countries:
        return "Top Market"
    else:
        return "Other Market"
df_clean["COUNTRY_BUCKET"] = df_clean["COUNTRY"].apply(country_bucket)
df_clean["COUNTRY_BUCKET"].value_counts()
```

```python
df_clean["DAYS_SINCE_LASTORDER"].describe()
```

```python
# Pick Thresholds
def recency_bucket(days):
    if days <= 365:
        return "Recent"   # ordered wuthin -1 year
    elif days <= 2*365:
        return "Warm"     # 1-2 years
    else:
        return "Dormant"  # > 2 years

df_clean["RECENCY_BUCKET"] = df_clean["DAYS_SINCE_LASTORDER"].apply(recency_bucket)
df_clean["RECENCY_BUCKET"].value_counts()
```

### Customer-Level Aggregated Features
### We"ll create features at customer level and merge them back:

1. CUST_TOTAL_SALES - Total revenue from that customer
2. CUST_NUM_ORDERS - Number of orders
3. CUST_AVG_SALES - Average order value

```python
cust_groups = df_clean.groupby("CUSTOMERNAME")["SALES"].agg(
    CUST_TOTAL_SALES = "sum",
    CUST_NUM_ORDERS = "count",
    CUST_AVG_SALES = "mean"
)
cust_groups.head()
```

```python
# Now merge back to each row
df_clean = df_clean.merge(cust_groups, on="CUSTOMERNAME", how="left")

df_clean[["CUSTOMERNAME", "SALES", "CUST_TOTAL_SALES",
          "CUST_NUM_ORDERS", "CUST_AVG_SALES"]].head()
```

```python
# YOU CAN BUCKET CUSTOMERS
def customer_value_bucket(x):
    if x < df_clean["CUST_TOTAL_SALES"].quantile(0.33):
        return "Low Value Customer"
    elif x < df_clean["CUST_TOTAL_SALES"].quantile(0.66):
        return "Medium Value Customer"
    else:
        return "High Value Customer"

df_clean["CUSTOMER_VALUE_SEGMENT"] = df_clean["CUST_TOTAL_SALES"].apply(customer_value_bucket)

df_clean["CUSTOMER_VALUE_SEGMENT"].value_counts()
        
```

#### WE CAN CREATE:
1. Discount amount: MSRP - PRICEEACH
2. Discount % : (MSRP - PRICEACH)/MSRP

```python
df_clean["DISCOUNT_AMT"] = df_clean["MSRP"] - df_clean["PRICEEACH"]
df_clean["DISCOUNT_PCT"] = df_clean["DISCOUNT_AMT"] / df_clean["MSRP"]

df_clean[["MSRP", "PRICEEACH", "DISCOUNT_AMT", "DISCOUNT_PCT"]].head()
```

```python
# ALSO BUCKET DISCOUNT LEVEL
def discount_bucket(p):
    if p <= 0.05:
        return "No / Low Discount"
    elif p <= 0.20:
        return "Moderate Discount"
    else:
        return "High Discount"
df_clean["DISCOUNT_BUCKET"] = df_clean["DISCOUNT_PCT"].apply(discount_bucket)
df_clean["DISCOUNT_BUCKET"].value_counts()
```

### DEAL SIZE ENCODING (NUMERIC)
1. DEALSIE is categorical : SMALL, MEDIUM< LARGE
2. For ML, we may want a numeric encoding t

```python
deal_map = {
    "Small" : 1,
    "Medium" : 2,
    "Large" : 3
}
df_clean["DEALSIZE_ENC"] = df_clean["DEALSIZE"].map(deal_map)
df_clean[["DEALSIZE", "DEALSIZE_ENC"]].head()
```
